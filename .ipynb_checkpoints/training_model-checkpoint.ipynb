```python
# Numerical Arrays
import numpy as np

# Neural Network Library
import tensorflow.keras as kr

# Data Frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
```

```python
# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
# import the dataset into a dataframe, referenced and adapted from:
# https://www.tensorflow.org/tutorials/load_data/csv
dataframe = pd.read_csv('./datasets/powerproduction.csv')

dataframe.shape
```

# Predicting Power output using Linear Regression

```python
# Add the dataframe values to an array
speed = np.array(dataframe[['speed']])
power = np.array(dataframe[['power']])

x = np.reshape(speed, (-1, 1))
y = np.reshape(power, (-1, 1))

# declare the model
model = kr.Sequential()

model.add(kr.layers.Flatten())
# give the model 1 neuron
model.add(kr.layers.Dense(15, input_shape=(1,), activation='linear', 
                          kernel_initializer='ones', 
                          bias_initializer='zeros'))

# compile the model
model.compile(optimizer="adam", loss='mean_squared_error', metrics=['accuracy'])
```

```python
# train the model
model.fit(x, y, epochs=500)
```

```python
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green")

# set the y_pred based on the speed the model receives
y_pred = model.predict(x)

plt.plot(x, y_pred, color='red')
plt.show()
```

```python
speed_val = input()
model.predict([float(speed_val)])
```

```python
model.evaluate(speed, power)
```

# Predicting Power output using Logistic Regression

```python
# Add the dataframe values to an array
speed = np.array(dataframe[['speed']])
power = np.array(dataframe[['power']])

x = np.reshape(speed, (-1, 1))
y = np.reshape(power, (-1, 1))

# create model
model = kr.Sequential()

model.add(kr.layers.Dense(10, input_dim=1, activation='relu'))
model.add(kr.layers.Dense(1, activation='sigmoid'))

model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(speed, power, epochs=500, validation_split=0.33, batch_size=100)
```

```python
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green")

# set the y_pred based on the speed the model receives
y_pred = model.predict(speed)

plt.plot(speed, y_pred, color='red')
```
