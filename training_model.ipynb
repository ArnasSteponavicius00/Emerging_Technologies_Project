```python
# Numerical Arrays
import numpy as np

# Neural Network Library
import tensorflow.keras as kr
from tensorflow.keras.layers import Dense

# Data Frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt

# Scikit for preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
# import the dataset into a dataframe, referenced and adapted from:
# https://www.tensorflow.org/tutorials/load_data/csv
dataframe = pd.read_csv('./datasets/powerproduction.csv')

dataframe.shape
```

# Predicting Power output using Linear Regression

```python
# Add the dataframe values to a numpy array
speed = dataframe['speed'].values
power = dataframe['power'].values

# Plot the dataframe
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="red")
plt.show()
```

```python
# preprocess the data
scaler = StandardScaler()

scaler.fit(dataframe)

scaler.transform(dataframe)
```

```python
# add the columns to a numpy array
speed = dataframe['speed'].values
power = dataframe['power'].values

# reshape the data
x = np.reshape(speed, (-1, 1))
y = np.reshape(power, (-1, 1))

# plot the speed and power
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="red")
plt.show()
```

```python
# declare the model
model = kr.Sequential()

model.add(kr.layers.Flatten())

# give the model 1 neuron
model.add(Dense(15, activation='relu', kernel_initializer="zeros", bias_initializer="ones"))

# compile the model
model.compile(optimizer="adam", loss='mean_squared_error', metrics=['accuracy'])
```

```python
# train the model
model.fit(x, y, epochs=500)
```

```python
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green", label='actual')

# set the y_pred based on the speed the model receives
y_pred = model.predict(x)

plt.plot(x, y_pred, color='red', label='prediction')
plt.show()
```

```python
speed_val = input()
model.predict([float(speed_val)])
```

```python
model.evaluate(speed, power)
```

# Predicting Power output using regression and multiple hidden layers

```python
# Add the dataframe values to an array
speed = np.array(dataframe[['speed']])
power = np.array(dataframe[['power']])

# create model
model = kr.Sequential()

# Add multiple hidden layers and used ReLU as activation method.
# reasons for using relu, is that sigmoid and tanh sometimes tend to make
# gradients vanish as they get close to 0 during backpropagation
model.add(Dense(128, input_dim=1, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# use the linear activation method in the final layer for regression
model.add(kr.layers.Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(speed, power, epochs=256)
```

```python
# Plot the dataframe
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green")

# set the y_pred based on the speed the model receives
y_pred = model.predict(x)

# plot the prediction
plt.plot(x, y_pred, color='red')
```

```python
value = input()
model.predict([float(value)])
```

```python
model.evaluate(x, y)
```
