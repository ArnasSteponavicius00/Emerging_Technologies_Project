```python
# Numerical Arrays
import numpy as np

# Neural Network Library
import tensorflow.keras as kr

# Data Frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
```

```python
f = lambda x: 3.0 * x + 1.0
```

```python
# import the dataset, referenced and adapted from:
# https://www.tensorflow.org/tutorials/load_data/csv
dataset = pd.read_csv('./datasets/powerproduction.csv',
                   names=['speed', 'power'],
                   skiprows=1)

# create the training dataframe
train_dataset = dataset.copy()

# create a test dataframe
test_dataset = dataset.copy()
```

```python
# Create a neural network with three neurons.
model = kr.models.Sequential()
model.add(kr.layers.Dense(3, input_shape=(1,), activation="linear", 
                          kernel_initializer='ones', 
                          bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')
```

```python
# Train the neural network on our training data.
model.fit(train_dataset['power'], train_dataset['speed'], epochs=500)
```

```python
model.predict([5.0, 20.0, 15.0, 25.0])
```

```python
plt.scatter(train_dataset['speed'], train_dataset['power'], label='actual', edgecolor='green')
plt.plot(train_dataset['speed'], model.predict(train_dataset['speed']), label='prediction', color='red')
plt.legend();
```
