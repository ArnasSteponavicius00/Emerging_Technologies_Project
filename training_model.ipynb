## Training a Neural Network to predict the [Power Output] of a wind turbine based on a given [Wind Speed]

### Imports

```python
# Numerical Arrays
import numpy as np

# Neural Network Library
import tensorflow.keras as kr
from tensorflow.keras.layers import Dense

# Data Frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt

# Scikit for preprocessing
from sklearn.preprocessing import StandardScaler
```

### Set the plotting style and size of the graph

```python
# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [12, 6]
```

Read in the data from the csv file.
The csv file consists of 500 rows and 2 columns as can be seen when we call the shape function on the dataframe.
The 2 columns are split into Speed and Power

```python
# import the dataset into a dataframe, referenced and adapted from:
# https://www.tensorflow.org/tutorials/load_data/csv
dataframe = pd.read_csv('./datasets/powerproduction.csv')

dataframe.shape
```

# Predicting Power output using Linear Regression

1. Get the columns from the dataframe, add the values into a numpy array and assign them to a variable
2. Plot the points on a scatter graph to visualise the data

```python
# Add the dataframe values to a numpy array
speed = dataframe['speed'].values
power = dataframe['power'].values

# Plot the dataframe
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="red")
plt.show()
```

### Creating the model:

Initially I am going to show off using linear regression to predict output.

1. Declare the model
2. Create a network with a single neuron
3. Compile the model using the ADAM optimizer and calculate the loss using mean squared error

```python
# declare the model
linear_model = kr.Sequential()

# Train the model using a linear activation function
linear_model.add(Dense(1, activation='linear', kernel_initializer="zeros", bias_initializer="ones"))
                       
# compile the model
linear_model.compile(optimizer="adam", loss='mean_squared_error', metrics=['accuracy'])
```

Train the model and run it for 500 epochs.
Epochs are the number of times the network iterates over the dataset it is passed both down and up the way ONCE.

```python
# train the model
linear_model.fit(speed, power, epochs=500)
```

### Plot the results

```python
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green", label='actual')

# set the y_pred based on the speed the model receives
y_pred = linear_model.predict(speed)

plt.plot(speed, y_pred, color='red', label='prediction')
plt.legend()
plt.show()
```

As you can see from the above graph, the neural network is able to predict values, however the linear regression algorithm might not be the best solution to this problem, reasons being.

- While the dataset does only contain two values (speed and power), when we initially created the scatter plot we could immediately tell that the plot did not create a linear pattern, but more of a sigmoid. Giving us the first hint that linear regression might not be the best way to fit our predictions to the graph

# Predicting Power output using regression and multiple hidden layers

In this case, we do the same thing as above when we trained the model using linear regression.

1. Add the columns to a numpy array and assign them variables
2. Declare a model
3. This time, we will use a different activation function "relu" which stands for Rectified Learning Unit. This function is now the standard and has replaced "sigmoid" activation functions. One of the reasons being.

    - Sigmoid function have a what is called "vanishing gradients" which is when a neuron gets to close to a zero value it has a chance of disappearing, meaning some of the weight values might not get updated.
    
    - ReLU eliminates the the vanishing gradient problem and is also zero-centered. The function should only be used in the hidden layers and activated using a linear activation when using it for regression.
    
#### References:
[Activation Functions](https://www.youtube.com/watch?v=-7scQpJT7uo)

```python
# Add the dataframe values to an array
speed = np.array(dataframe[['speed']])
power = np.array(dataframe[['power']])

# create model
model = kr.Sequential()

# Add multiple hidden layers and used ReLU as activation method.
# reasons for using relu, is that sigmoid and tanh sometimes tend to make
# gradients vanish as they get close to 0 during backpropagation
# also, not zero centered, making all weights start at at 0 and end positive, not allowing for
# negative weights
model.add(Dense(128, input_dim=1, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# use the linear activation method in the final layer for regression
model.add(kr.layers.Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

### Train the model

```python
# Train the model
model.fit(speed, power, epochs=500)
```

### Plot the prediction

```python
# Plot the dataframe
dataframe.plot(kind='scatter', x='speed', y='power', title='Wind Turbine Output', edgecolor="green", label="actual")

# set the y_pred based on the speed the model receives
y_pred = model.predict(speed)

# plot the prediction
plt.plot(speed, y_pred, color='red', label="prediction")
plt.legend()
plt.show()
```

### Prediction

From the graph above, we can see that the model now fits the prediction to the values alot more closely. We can also see that the prediction line started dropping near the end, which could be attributed to possible reasons like, possibly the turbine shutting down when speeds reach a certain speed, Or maybe there was a sensor failure which failed to return the wind speed at the time

Simple input box to check prediction values

```python
value = input()
model.predict([float(value)])
```

Evaluate the models accuracy and loss

```python
model.evaluate(speed, power)
```
